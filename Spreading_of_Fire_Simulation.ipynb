```python
import random

# Create burning function
def isBurn(chance):
  threshold = random.randint(1, 10)
  if (threshold <= (chance/10)):
    return 1
  else:
    return 0
```

```python
import pandas as pd
import numpy as np

# Create an array of a burning tree P with initial value (7,7) (the tree is in the center of area L)
P = [(8, 8)]
#NumOfBurningTree = 1

# Initialize necessary variables
burnProbability = int(input('Input Burn Probability (%): '))
iter = 0
```

```python
# Create a spreading for 17 iterations
for i in range(1, 18):

  iter = iter + 1
  newP = []

  for p in P:

    # Check the surrounding 4 trees
    # for the tree in the south side
    if (p[0] - 1 >= 0):
      if isBurn(burnProbability):
        if not (p[0] - 1, p[1]) in P:
          newP.append((p[0] - 1, p[1]))
    # for the tree in the west side
    if (p[1] - 1 >= 0):
      if isBurn(burnProbability):
        if not (p[0], p[1] - 1) in P:
          newP.append((p[0], p[1] - 1))
    # for the tree in the north side
    if (p[0] + 1 <= 16):
      if isBurn(burnProbability):
        if not (p[0] + 1, p[1]) in P:
          newP.append((p[0] + 1, p[1]))
    # for the tree in the east side
    if (p[1] + 1 <= 16):
     if isBurn(burnProbability):
        if not (p[0], p[1] + 1) in P:
          newP.append((p[0], p[1] + 1))

  for newp in newP:
    if not newp in P:
      P.append(newp)

  # Conclusion after the 'iter'-th iteration
  print("After", iter, "iterations", ": Percentage of burning trees = (",(len(P)/(17*17))*100,")")

  # Create a matrix of area L with the size of 17x17, note: 0 = tree, 1 = burning tree
  # (because every cell in the matrix is filled with tree/burning tree)
  L = np.zeros([17, 17], dtype = int)

  for p in P:
    x = p[0]
    y = p[1]
    L[x][y] = 1

  import matplotlib.pyplot as plt
  plt.axis([0, 16, 0, 16])
  plt.imshow(L)
  plt.show()
```
